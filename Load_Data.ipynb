```python
from sklearn import datasets

digits = datasets.load_diabetes()
features = digits.data
target = digits.target

features[0]
```

```python
from sklearn.datasets import make_regression
fatures,target,coefficients = make_regression(n_samples=100,
                                             n_features=3,
                                             n_informative=3,
                                             n_targets=1,
                                             noise=0.0,
                                             coef=True,
                                             random_state=1)

print('Features Matrix\n',features[:3])
print('Target Vector\n',target[:3])
```

```python
from sklearn.datasets import make_classification
fatures,target = make_classification(n_samples=100,
                                             n_features=3,
                                             n_informative=3,
                                             n_redundant=0,
                                             n_classes=2,
                                             weights=[.25,.75],
                                             random_state=1)

print('Features Matrix\n',features[:3])
print('Target Vector\n',target[:3])
```
